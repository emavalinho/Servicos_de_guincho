Reboques/Guinchos de Curitiba do site do IPPUC e faça um mapa com os bairros, os pontos, o mapa de calor e os clusters. Pode ser feito também com dados do OpenStreetMap, baixados no QGIS com o complemento QuickOSM.
https://ippuc.org.br/geodownloads/geo.htm

```
#Exercício 3B
from google.colab import drive
drive.mount ('/content/drive')

! pip install geopandas

import folium
from folium import plugins
from folium.plugins import MeasureControl
from folium.plugins import MarkerCluster
from folium.plugins import HeatMap
import geopandas as gpd
```

```
#mapa somente com os pontos
SERVICOS_DE_REBOQUE = '/content/drive/My Drive/DESENV_APP_GEOESPACIAIS/SERVICOS_REBOQUE.geojson' # Caminho corrigido para o arquivo GeoJSON

m = folium.Map (location = [-25.5,-49.3],
               tiles = 'OpenStreetMap',
               zoom_start =  12)

folium.GeoJson(
   SERVICOS_DE_REBOQUE,
   name='SERVICOS_DE_REBOQUE',
   ).add_to(m)

m
```

```
# mapa com os bairros, cluster e mapa de calor e com as funções de liga/desliga camada e medição de distância
# Corrigido o caminho para 'bairros.geojson' para 'Divisas_de_bairros.geojson'. Por favor, verifique se 'Divisas_de_bairros.geojson' está realmente nesta pasta.
bairros = '/content/drive/My Drive/DESENV_APP_GEOESPACIAIS/Divisas_de_bairros.geojson'
style = {'fillcolor' : '#F5DEB3' , 'color' : '#8B0000'}

m = folium.Map (location = [-25.5,-49.3],
               tiles = 'openstreetmap',
               zoom_start =  12)

folium.GeoJson(
   bairros,
   name='bairros',
   tooltip=folium.GeoJsonTooltip(fields=['NOME', 'FONTE']),
   style_function=lambda x: style).add_to(m)

# Corrigido o caminho para 'SERVICOS_DE_REBOQUE.geojson' para corresponder à célula anterior
SERVICOS_DE_REBOQUE = '/content/drive/My Drive/DESENV_APP_GEOESPACIAIS/SERVICOS_REBOQUE.geojson'
# Lendo os dados de reboque. A variável 'creches' global foi substituída por 'reboque' definida localmente.
df = gpd.read_file(SERVICOS_DE_REBOQUE)

locations = []

for idx, row in df.iterrows():
    locations.append([row['geometry'].y, row['geometry'].x])

m.add_children(MarkerCluster(locations=locations, name = 'Creches Municipais')) # 'Creches Municipais' pode precisar ser ajustado se os dados forem de reboque

for idx, row in df.iterrows():
    locations.append([row['geometry'].y, row['geometry'].x])

HeatMap(locations,name = 'Mapa de Calor').add_to(m)


folium.LayerControl().add_to(m)
# Corrigido o método para adicionar o controle de medição.
m.add_child(MeasureControl())

m
```
